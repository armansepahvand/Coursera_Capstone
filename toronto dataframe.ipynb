```python
with open ('Toronto_data.html') as Thtml:
    Tsoup=BeautifulSoup(Thtml, 'lxml')
```

```python
table=Tsoup.find('table')
```

```python
table=Tsoup.tbody
len(table)
```

```python
rows=Tsoup.table.tbody.find_all('tr')
```

```python
data=[]
master=[]
```

```python
for row in rows:
    for j in row.find_all('td'):
        data.append(j.text)
    master.append(data)
    data=[]
    
```

```python
for row in master:
    for i,s in enumerate(row):
        row[i]=s.strip()
        
```

```python
df=pd.DataFrame(master)
```

```python
df.columns=['PostalCode', 'Borough', 'Neighborhood']
```

```python
df1=df.replace('', np.nan)
df1.dropna(how='any', inplace=True)
```

```python
df2=df1.reset_index(drop=True)
```

```python
df2.replace(' / ', ' , ', regex=True, inplace=True)
```

```python
dfsorted=df2.sort_values(by=['PostalCode'])
```

```python
dfsorted.reset_index(drop=True)
```
